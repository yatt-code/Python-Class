#Tuple

###List

*   Mutable: You can change the elements (add, remove, or modify)
    after the list is created.
*   Syntax: Defined using square brackets []. python my_list = [1,
    2, 3]
*   Performance: Slightly slower due to the flexibility of
    mutability.
*   Use Case: Best when you need a collection of items that may
    change over time.


###Tuple

*   Immutable: Once created, you cannot change the elements (no
    adding, removing, or modifying).
*   Syntax: Defined using parentheses (). python my_tuple = (1, 2,  
    3)
*   Performance: Faster than lists due to immutability.
*   Use Case: Best when you need a collection of items that should
    not change, ensuring data integrity.
*   Tuple is slightly faster at processing compared to list

Key Differences Between 'return' and 'print'

*  Functionality:

      * return sends a value back to the caller and terminates the
       function execution.
      * print outputs a value to the console but does not affect function execution or return values.
* Context:

      * return is used inside functions to send a result back.
      * print can be used anywhere to display information to the
        user.
* Effect on Function:

      * return can determine the outcome of a function and can be used to pass data between functions.
      * print is primarily used for debugging or displaying information and has no effect on function results or data flow.

*print vs return*

To summarize, "print" is used to output values to the console for debugging or informational purposes, while "return" is used to send a value back from a function to the calling code. If you want to use the result of a function in other parts of your code, you need to use "return" to make that value available outside the function.



*print just shows the human user a string representing what is going on inside the computer. The computer cannot make use of that printing. return is how a function gives back a value. This value is often unseen by the human user, but it can be used by the computer in furtherfunctions.*

#Remove duplicate

```python
#write a function to remove duplicates from a list of numbers

def remove_duplicate(list_duplicate):
  list = []
  for i in list_duplicate:
    if i not in list:
      list.append(i)
  return list
```

```python
list_duplicate = [100, 20, -30, 100, 20, 0, -30]

list = remove_duplicate(list_duplicate)
print (list)
```

```python
import geopandas as gpd

# Load a world map using GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a geographic heatmap
world.plot(column='gdp_md_est', cmap='OrRd', legend=True)
plt.title('Geographic Heatmap of GDP')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
x = 0.1 + np.arange(8)
y = [1.0, 7.5, 4.5, 5.6, 5.2, 3.6, 2.3, 4.4]
colors = ['green', 'yellow', 'red', 'orange', 'black', 'blue', 'purple', 'pink']

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7, color=colors)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#create a bar plot

plt.figure(figsize=(8, 6))
plt.bar(x, y, color= 'skyblue')


plt.xlabel('Price')
plt.ylabel('Apple sold')
plt.title('Apple in Market A')
plt.legend(['Apple'])

plt.show()
```
