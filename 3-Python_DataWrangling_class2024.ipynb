# Deal with Missing Value

```python
mean_norm = mydata['normalized-losses'].mean()

mean_norm
```

normalized- losses type is object, need to change type to "float"

recommended to change to float because if float changed to int, some data will be missing

###Change object to float

```python
mean_norm = mydata['normalized-losses'].astype(float).mean()

mean_norm
```

###Replace the missing value

```python
#replace (nan, mean)

mydata['normalized-losses'].replace(np.nan, mean_norm, inplace = True)

#Checking total number of missing value in a column AGAIN
mydata.isnull().sum()
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace = True)

#Checking total number of missing value the bore column AGAIN
mydata.isnull().sum()
```

```python
# Finding the mean value of the 'stroke' column
mean_stroke= mydata['stroke'].astype(float).mean()
mean_horsepower= mydata['horsepower'].astype(float).mean()
mean_peak= mydata['peak-rpm'].astype(float).mean()

# Replace the missing value in 'stroke' with the mean value
mydata['stroke'].replace(np.nan, mean_stroke, inplace = True)
mydata['horsepower'].replace(np.nan, mean_horsepower, inplace = True)
mydata['peak-rpm'].replace(np.nan, mean_peak, inplace = True)

#Checking total number of missing value the stroke column AGAIN
mydata.isnull().sum()
```

```python
#Here we are going to drop the missing value in price

mydata.dropna(subset=['price'],axis=0, inplace=True)

#Checking total number of missing value the price column AGAIN
mydata.isnull().sum()
```

#**Data Normalization**

nak bagi data tu tak berapa jauh

```python
mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

##Simple feature scaling

```python
#simple feature scaling. nak 0-1

mydata['width'] = mydata['width']/mydata['width'].max()

mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'].describe()
```

##Min-max

```python
#2-min-max

mydata['length'] = (mydata['length']-mydata['length'].min())/(mydata['length'].max()-mydata['length'].min())

mydata[['width', 'length', 'height']].head()
```

```python
mydata['length'].describe()
```

##Z-Score

```python
#3- Z-score

mydata['height'] = (mydata['height']-mydata['height'].mean())/mydata['height'].std()

mydata[['width', 'length', 'height']].head()
```

```python
#use the same method for all feature for the normalization. cannot mix. can choose any method, just cannot mix

mydata[['width', 'length', 'height']].describe()
```

#**Data Binning**

```python
mydata['price'].describe()
```

```python
#divide price to 3 categpries

price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)
print(price_bins)

price_groups = ['Low Price', 'Mid Price', 'High Price']

mydata_new1['price_bins'] = pd.cut(mydata_new1['price'], price_bins, labels = price_groups, include_lowest=True)
mydata_new1[['price','price_bins']].head(10)
```
